```python
from pyspark.sql.functions import *

cad_df = spark.read.format("delta").load("/mnt/delta/bronze/cad_metadata")

cad_silver = cad_df \
.withColumn("created_date",to_date("created_date")) \
.dropDuplicates(["design_id","version"])

cad_silver.write.format("delta") \
.mode("overwrite") \
.save("/mnt/delta/silver/cad_metadata")
```
